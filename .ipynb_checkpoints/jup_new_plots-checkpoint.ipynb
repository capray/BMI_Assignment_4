```python
# load the libraries using the normal convention
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# prevent plotting errors
%matplotlib inline
patient_23234_xls = pd.ExcelFile('patient_23234_izzy.xlsx')

patient_23234_plate1= patient_23234_xls.parse(0, skiprows=1)

patient_23234_plate1
```

```python
patient_23234_plate1
```

```python
V1 = patient_23234_plate1.drop(patient_23234_plate1.index[[4, 5, 6, 7]])
V1
```

```python
V2 = patient_23234_plate1.drop(patient_23234_plate1.index[[0, 1, 2, 3]])
V2
```

```python
from functools import reduce
import operator
patient = V1['Sample ID'].iloc[0]
patient_age = V1['Age'].iloc[0]
patient_gender = V1['Gender'].iloc[0]
patient_hospital = V1['Hospital '].iloc[0]

pt_info = [str(patient), '(', patient_gender, str(patient_age), 'yr', patient_hospital, ') PSMalpha2']
pt_info_title = ' '.join(pt_info)
pt_info_title
```

```python
plt.figure(1); # plotting figure 1
width = 6  # setting width
height = 6  # setting height
plt.figure(figsize=(width, height))
plt.suptitle(pt_info_title, fontsize=14, fontweight='bold')
visit1, = plt.loglog(V1["Dilution"],V1["PSMalpha2"], color='r', marker = 'o', markerfacecolor= 'none', markersize= 10, markeredgewidth= 2, label= 'Visit 1')
visit2, = plt.loglog(V2["Dilution"],V2["PSMalpha2"], color='b', marker = 'o', markerfacecolor= 'none', markersize= 10, markeredgewidth= 2, label= 'Visit 2')
plt.xlabel('Dilution', fontsize=14)
plt.ylabel('Intensity', fontsize=14)
plt.legend(handles=[visit1, visit2], loc=1)


plt.show()
```
